## 3378. Count connected components in LCM graph
You are given an array of positive integers `nums`. Consider the following graph:
- There are `nums.length` nodes, the `i`-th node (0-indexed) has a value of `nums[i]`.
- There is an undirected edge between the `i`-th and `j`-th nodes if and only if `lcm(nums[i], nums[j]) <= k`, where `lcm(x, y)` is the least common multiple of `x` and `y`.
Return the number of connected components in the graph.
### Example 1:
```
Input: nums = [4, 7, 15, 8], k = 5
Output: 4
Explanation: The graph has 4 connected components: [4], [7], [15], and [8].
```
### Example 2:
```
Input: nums = [1, 2, 3, 4], k = 4
Output: 2
Explanation: The graph has 2 connected components: [1, 2, 3, 4].
```
### Constraints:
- `1 <= nums.length <= 10^4`
- `1 <= nums[i] <= 10^4`
- `1 <= k <= 10^4` 

````python
class UnionFind:
    def __init__(self, arr):
        self.parent = arr
        self.rank = [0]*len(arr)
        self.component_count = 0
    
    def find(self, x):
        if self.parent[x] != x:
             self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)

        if self.rank[root_x] > self.rank[root_y]:
            self.parent[root_y] = root_x
        elif self.rank[root_x] < self.rank[root_y]:
            self.parent[root_x] = root_y
        else:
            self.parent[root_y] = root_x
            self.rank[root_x] += 1

        return True
        
````
