## Union-Find Data Structure

The Union-Find data structure, also known as Disjoint Set Union (DSU), is a powerful tool for managing and merging disjoint sets. It provides efficient operations for union and find, making it ideal for solving problems related to connectivity and grouping.

- **MakeSet(x):** create the singleton set {x}
- **Union(x, y):** merge $S_{x}$ and $S_{y}$ into a single set, where $S_{x}$ and $S_{y}$ are the sets containing x and y, respectively. If $S_{x}$ and $S_{y}$ are already in the same set, do nothing.
- **Find(x):** return the representative (or "root") of the set containing x. This can be used to determine if two elements are in the same set.

## 1. Initialization MakeSet(x)

First we need to create the class and initialize our data. For every node, we treat it as a separate set, so we can initialize the parent of each node to itself and the rank (or size) of each set to 0.
```Running time: O(n) for initializing n elements.```


## 2. The Find Operation (with Path Compression)

The `Find` operation is used to determine which set a particular element belongs to. It returns the representative (or "root") of the set containing the element. To optimize this operation, we can use path compression, which flattens the structure of the tree whenever `Find` is called, making future operations faster.
```Running time: near O(1) on average, O(log n) in the worst case.```


## 3. The Union Operation (with Union by Rank)

When merging two sets, we find their respective roots. If they have different roots, we attach the shorter tree under the root of the taller tree using our rank array. This prevents the trees from becoming long, linked lists.  
```Running time: near O(1) on average, O(log n) in the worst case.```

Rank can be defined as the height of the tree or the size of the set. In this implementation, we will use rank as the height of the tree.


````python
class UnionFind:
    def __init__(self, size):
        # initially each node is its own parent (singleton set)
        self.parent = [i for i in range(size)]
        # rank is used to keep the tree flat during union operations
        self.rank = [0]*size

    def find(self, x):
        # if x is not the root, recursively find the root and update the parent
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x]) # Path compression
        return self.parent[x]
    
    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)

        # if they are already in the same set, do nothing 
        if root_x == root_y:
            return False
        
        # Attach the smaller rank tree under the root of the higher rank tree
        if self.rank[root_x] > self.rank[root_y]:
            self.parent[root_y] = root_x
        elif self.rank[root_x] < self.rank[root_y]:
            self.parent[root_x] = root_y
        else:
            # if they are the same, pick one to be the root and increment its rank 
            self.parent[root_y] = root_x
            self.rank[root_x] += 1

        return True 
````

## 4. Graph Connectivity (Application of Union-Find)

To determine if two nodes in a graph are connected, we can use the Union-Find data structure. We can iterate through the edges of the graph and perform union operations for each edge. After processing all edges, we can check if two nodes belong to the same set by using the find operation.
```Running time: O(E * α(n)) for processing E edges, where α(n) is the inverse Ackermann function, which grows very slowly and is practically constant for all reasonable values of n.```

````python
def connected_components(num_vertices, edges):
    # for all v in V(G) do MakeSet(v)
    uf = UnionFind(num_vertices)

    # for all (u,v) do
    for u, v in edges:
        # if find(u) != Find(v) then Union(u,v)
        uf.union(u, v)
    
    return uf


def is_connected(uf, u, v):
    # return True if they share the same representative root
    return uf.find(u) == uf.find(v)


# Example usage 
# let's say we have 5 vertices (0 to 4) and some edges 
vertices = 5
edges = [(0,1), (1,2), (3,4)]

# Preprocessing 
uf_graph = connected_components(vertices, edges)


# Example queries 
print(f"Are 0 and 2 connected? {is_connected(uf_graph, 0, 2)}")  
print(f"Are 0 and 3 connected? {is_connected(uf_graph, 0, 3)}")
    
````
