## Connected Component Counter 
Given an undirected graph, count the number of connected components in the graph  
Example 1:
```
Input: n = 5, edges = [[0,1],[1,2],[3,4]]
Output: 2
Explanation: There are two connected components: 0, 1, 2 and 3, 4.

```
Example 2:
```
Input: n = 5, edges = [[0,1],[1,2],[2,3],[3,4]]
Output: 1
Explanation: There is one connected component: 0, 1, 2, 3 and 4.
```


Logic: At very beginning we have n connected components.  
Every time we connect two nodes, we reduce the number of connected components by 1.  
if we connect two nodes that are already in the same connected component, we do not reduce the number of connected components.

````python
class UnionFind:
    def __init__(self, size):
        # initially each node is its own parent (singleton set)
        self.parent = [i for i in range(size)]
        # rank is used to keep the tree flat during union operations
        self.rank = [0]*size

        # start with n number of disconnected components 
        self.component_count = size

    def find(self, x):
        # if x is not the root, recursively find the root and update the parent
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x]) # Path compression
        return self.parent[x]
    
    def union(self, x, y):
        root_x = self.find(x)
        root_y = self.find(y)

        # if they are already in the same set, do nothing 
        if root_x == root_y:
            return False
        
        # Attach the smaller rank tree under the root of the higher rank tree
        if self.rank[root_x] > self.rank[root_y]:
            self.parent[root_y] = root_x
        elif self.rank[root_x] < self.rank[root_y]:
            self.parent[root_x] = root_y
        else:
            # if they are the same, pick one to be the root and increment its rank 
            self.parent[root_y] = root_x
            self.rank[root_x] += 1
        
        # after union of two distinct components, we reduce the total by 1
        self.component_count -= 1

        return True 

    def get_count(self):
        return self.component_count
````

````python
# Test examples 
vertices = 6 

# Edges forming two separate groups: (0-1-2) and (3-4). Node 5 is completely alone!
edges = [(0, 1), (1, 2), (0,2), (3, 4)]

uf_graph = UnionFind(vertices)

for u, v in edges:
    uf_graph.union(u, v)

print(f"Total distinct components remaining: {uf_graph.get_count()}")
````

## Complexity Analysis

Let's define the variables:
- V: the number of vertices in the graph
- E: the number of edges in the graph

### Space Complexity: O(V) for the Union-Find data structure, which stores the parent and rank of each vertex.  
- We allocate parent and rank arrays of size V to keep track of the connected components.
- therefore, the space complexity is bounded by `O(V)`.

### Time Complexity 

```__init__```: we iterate through all vertices to initialize the parent and rank arrays, which takes O(V) time.  

```find```: with path compression, the time complexity is nearly O(1) for each operation, but more precisely, it is O(α(V)), where α is the inverse Ackermann function.  

```union```: with union by rank, the time complexity is also O(α(V)) for each operation.  

Overall, for E edges, the total time complexity is O(E * α(V)).  

Overall, the time complexity of the algorithm is O(V + E * α(V)).

And Space Complexity is O(V) due to the Union-Find data structure.  

Inverse Ackermann function, α(n), grows very slowly and is less than 5 for all practical values of n, making it effectively constant time for most applications.
